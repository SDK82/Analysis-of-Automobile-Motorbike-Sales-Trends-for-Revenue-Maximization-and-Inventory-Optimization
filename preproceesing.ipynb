## Transform dataframes with only required columns :

```python
dfs = [df_20_21,df_21_22,df_22_23,df_23_24]
```

```python
required_cols = ['bill_no', 'date', 'c_code', 'client', 'name', 'address',
                     'it_code', 'group', 'brand', 'model', 'col', 'engine',
                     'amount1', 'cg', 'cgst', 'sg', 'sgst', 'amount']
```

```python
def create_final_dataframe(raw_df):
    required_cols = ['bill_no', 'date', 'c_code', 'client', 'name', 'address',
                     'it_code', 'group', 'brand', 'model', 'col', 'engine',
                     'amount1', 'cg', 'cgst', 'sg', 'sgst', 'amount']
    final_df = raw_df[required_cols].copy()

    return final_df
```

```python
f_df_20_21 = create_final_dataframe(df_20_21)
f_df_21_22 = create_final_dataframe(df_21_22)
f_df_22_23 = create_final_dataframe(df_22_23)
f_df_23_24 = create_final_dataframe(df_23_24)
```

## Merge all dataframes

```python
def concatenate_dataframes(dataframes,axis=0,ignore_index=True):
    master_df = pd.concat(dataframes,axis=axis,ignore_index=ignore_index)
    return master_df
```

```python
list_of_f_dfs = [f_df_20_21,f_df_21_22,f_df_22_23,f_df_23_24]
```

```python
master_dataframe = concatenate_dataframes(list_of_f_dfs)
```

## Remove Surnames from dataframes

```python
def remove_surnames(full_name):
    if pd.notna(full_name):             # split() method is only called if the value is not NaN
        return full_name.split()[0]
    else:
        return full_name
```

```python

master_dataframe['name'] = master_dataframe['name'].apply(remove_surnames)
```

We collected the profit percentage from organization :

> Moped = 4%

> Motorcycle = 6%

> Scooter = 5%

> Scooty = 5%

> So, calculated the cost amount and create a new columnn on the master
dataframe:











```python
def create_profit_percentage(df):
    df['group']=df['group'].upper()
    profit_percentage = 0

    if df['group']=='MOPED':
         return '4%'
    elif df['group']=='MOTORCYCLE':
        return'6%'
    elif df['group']=='SCOOTER' or df['group']=='SCOOTY':
        return '5%'

```

```python
master_dataframe['profit'] = master_dataframe.apply(create_profit_percentage, axis=1)
```

```python
master_dataframe
```

```python
def calculate_cost_amount(df):

    df['group']=df['group'].upper()
    profit_percentage = 0

    if df['group']=='MOPED':
        profit_percentage = 4
    elif df['group']=='MOTORCYCLE':
        profit_percentage = 6
    elif df['group']=='SCOOTER' or df['group']=='SCOOTY':
        profit_percentage = 5


    original_amount = df['amount1']/(1+(profit_percentage/100))
    return original_amount
```

```python
master_dataframe['Original Amount'] = master_dataframe.apply(calculate_cost_amount, axis=1)
```
